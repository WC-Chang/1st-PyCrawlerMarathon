```python
# 下載檔案
import urllib.request
import zipfile

res = "http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314"
urllib.request.urlretrieve(res, "./data/exampleDay3.zip")
f = zipfile.ZipFile('./data/exampleDay3.zip')
f.extractall('./data')
```

```python
import os, sys

# 打开文件
dirs = os.listdir( './data' )

# 输出所有文件和文件夹
for file in dirs:
    print(file)
```

```python
# 讀檔案
fh = open("./data/64_72hr_CH.xml", "r",encoding="utf-8")
xml = fh.read()
fh.close()

print(xml)
```

```python
# 解析檔案內容
import xmltodict as xml
with open("./data/64_72hr_CH.xml",encoding="utf-8") as df:
    x = dict(xml.parse(df))
    print(x)
```
